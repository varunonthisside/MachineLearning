```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# load data
data = pd.read_csv('/content/drive/MyDrive/ML Projects Datasets/CustomerSegmentation/Mall_Customers.csv')
```

```python
#check for null values
data.isnull().sum()
```

```python
X_gender = pd.get_dummies(data.Gender,drop_first=True)
X_gender
```

```python
#choosing columns for clustering. Considering all except customer id
X = pd.DataFrame(data.iloc[:,2:5].values,columns=['age','annual_income','spending_score'])
X = pd.concat([X_gender,X],axis = 1)
X = np.asarray(X)
print(X)
```

```python
#choosing number of clusters
#WCSS

wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init = 'k-means++', random_state = 42)
  kmeans.fit(X)

  wcss.append(kmeans.inertia_)
```

```python
#plot an elbow graph

sns.set()
plt.plot(range(1,11),wcss)
plt.title('Elbow point graph')
plt.xlabel('clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# 8 clusters
# training the k-means clsutering model

kmeans = KMeans(n_clusters=5, init='k-means++', random_state=28)

# return a label for each data point based on their cluster
Y = kmeans.fit_predict(X)

print(Y)
```

```python
# Clusters - 0,1,2,3,4,5,6,7
# plotting all clusters and their centroids

plt.figure(figsize= (10,10))
plt.scatter(X[Y==0,2], X[Y==0,3], s = 50, c = 'red', label = 'Cluster1')
plt.scatter(X[Y==1,2], X[Y==1,3], s = 50, c = 'blue', label = 'Cluster2')
plt.scatter(X[Y==2,2], X[Y==2,3], s = 50, c = 'green', label = 'Cluster3')
plt.scatter(X[Y==3,2], X[Y==3,3], s = 50, c = 'yellow', label = 'Cluster4')
plt.scatter(X[Y==4,2], X[Y==4,3], s = 50, c = 'pink', label = 'Cluster5')
#plt.scatter(X[Y==5,2], X[Y==5,3], s = 50, c = 'pink', label = 'Cluster6')
#plt.scatter(X[Y==6,2], X[Y==6,3], s = 50, c = 'purple', label = 'Cluster7')
#plt.scatter(X[Y==7,2], X[Y==7,3], s = 50, c = 'orange', label = 'Cluster8')

plt.scatter(kmeans.cluster_centers_[:,2], kmeans.cluster_centers_[:,3], s = 100, c = 'cyan', label = 'Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
